```python
#Question 6
x = int(input())
if x == 0 or x == 1:
  print("True")
if x == 2 or x == 3:
  print("False")
if x % 2 == 0 or x % 3 == 0:
  print("False")

i = 5
while i*i<=x:
    if x % i == 0 or x % (i + 2) == 0:
        print("False")
    i += 6

print('True')
```

```python
#Question 7
arr = [x for x in range(10,101)]
i  = 0
while i < len(arr):
  if arr[i] % 2 == 0 or arr[i] % 3 == 0 or arr[i] %5 == 0 or arr[i] % 7 == 0:
    arr.pop(i)
  else:
    i += 1
print(arr)
```
