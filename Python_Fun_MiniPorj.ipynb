```python
#Question 1
miles = float(input("How many Miles you want to Travel : "))
if miles < 3:
  print("I suggest to travel with Bicycle")
elif miles < 300:
  print("I suggest to travel with Motorbike")
else:
  print("I suggest to travel with SuperCar")
```

```python
#Question 2
charges = float(input("Enter Charges per hour : "))
print(f"Charge per Day {charges * 24}")
print(f"Charge per Week {charges * 24 * 7}")
print(f"Charge per Month {charges * 30}")
budget = float(input("Enter Budget : "))
print(f"With budget of {budget} server can run for {budget / (charges * 24)}")
```
